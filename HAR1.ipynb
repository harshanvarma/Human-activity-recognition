### Unique values

```python
for column in data.columns:
    unique_values = data[column].unique()
    print(f"Unique values in {column}:\n", unique_values[:10])
    print(f"Total unique values in {column}: {len(unique_values)}\n")
```

### Counting Activity Occurrences

```python
activity_counts = data['Activity'].value_counts()
print("Activity counts:\n", activity_counts)
```

### Outlier Detection in Numeric Columns

```python
numeric_columns = data.select_dtypes(include=[np.number]).columns
for col in numeric_columns:
    Q1 = data[col].quantile(0.25)
    Q3 = data[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = data[(data[col] < lower_bound) | (data[col] > upper_bound)]
    print(f"Outliers in {col}: {len(outliers)}")
```

### Standard Deviation and Variance of Acceleration Columns

```python
acceleration_cols = ['Acceleration x (m/s^2)', 'Acceleration y (m/s^2)', 'Acceleration z (m/s^2)']
acceleration_std_dev = data[acceleration_cols].std()
acceleration_variance = data[acceleration_cols].var()

print("Standard Deviation of Acceleration Columns:\n", acceleration_std_dev)
print("Variance of Acceleration Columns:\n", acceleration_variance)
```

### Standard Deviation and Variance of Gforce columns

### Standard Deviation and Variance of gyroscope columns

### Calculating Total Acceleration Magnitude

```python
data['Total Acceleration Magnitude'] = np.sqrt(
    data['Acceleration x (m/s^2)']**2 +
    data['Acceleration y (m/s^2)']**2 +
    data['Acceleration z (m/s^2)']**2
)
print("Added 'Total Acceleration Magnitude' column:\n", data[['Total Acceleration Magnitude']].head())
```
